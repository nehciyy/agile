```python
import pandas as pd # data processing. 
import re
from bs4 import *
import urllib3
import nltk
nltk.download('omw-1.4')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer    
import requests   
```

```python
# This is the web page we will pull data 
url = "https://skillcrush.com/blog/41-tech-job-titles/"

# Pull the date from the web page
http = urllib3.PoolManager()
response = http.request('GET', url)

# Create a Beautiful Soup object, passing in the page data
soup = BeautifulSoup(response.data, "lxml")
```

```python

# Initialize lists to store data
headers = []
third_paragraphs = []

# Find all h3 elements
h3_elements = soup.find_all('h3')

# Get the list of English stopwords
stop_words = set(stopwords.words("english"))

# Initialize WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()

# Loop through h3 elements and find the third paragraphs after each h3
for h3 in h3_elements:
    # Find all paragraphs that are siblings of the current h3
    paragraphs = h3.find_next_siblings('p')
    
    # Check if there are at least 3 paragraphs after the h3
    if len(paragraphs) >= 3:
        third_paragraph = paragraphs[2].text
        
        # Remove number and dot from the front of h3 text
        header_text = h3.text.lstrip('1234567890. ')
        
        # Remove the first word "Skills:" from the beginning of the third paragraph
        third_paragraph = ' '.join(third_paragraph.split()[1:])
        
        # Tokenize and remove stopwords from third paragraph
        third_paragraph_tokens = word_tokenize(third_paragraph)
        third_paragraph_tokens = [lemmatizer.lemmatize(word) for word in third_paragraph_tokens if word.lower() not in stop_words]
        
        headers.append(header_text.strip())  # Remove leading/trailing spaces
        third_paragraphs.append(' '.join(third_paragraph_tokens))

# Create a DataFrame from the collected data
data = {'Header': headers, 'Skills': third_paragraphs}
df = pd.DataFrame(data)

    
```
